```python
pop = pd.read_excel('C:/Users/96650/OneDrive/Desktop/spiced_projects/sandbox/week_1/data/gapminder_population.xlsx', index_col=0)
```

```python
#pop.dropna(inplace=True)
pop.index
pop.index.name = 'country'
#pop.rename({'Total population': 'country'}, axis = 1, inplace = True)
```

```python
pop = pop.reset_index()
melt_pop = pop.melt(id_vars='country', var_name='year', value_name='population')
```

```python
melt_pop.head()
```

```python
fertility_pop = pd.merge(left=melt_fertility, right=melt_pop, on=['country', 'year'], how='left')
```

```python
life_fertility_pop = pd.merge(left=melt_life, right=fertility_pop, on=['country', 'year'], how='left')
```

```python
life_fertility_pop.head()
```

```python

life_fertility_pop.dropna(inplace=True)
```

```python
sns.heatmap(life_fertility_pop.isna())
```

```python
life_fertility_pop.head()
```

```python
df_subset = life_fertility_pop.loc[life_fertility_pop['year'] == 2000]
sns.scatterplot(x='life_rate', y='fertility_rate',
            data=df_subset, alpha=0.6)
```

```python
sns.set_style('ticks')

years = range(1960, 2000)


for year in years:
    
    plt.figure()
    plt.clf()
    df_filter = life_fertility_pop[life_fertility_pop['year'] == year]
    
    
    size = df_filter['population']

    
    scatterplot = sns.scatterplot(data=df_filter, x='life_rate', y='fertility_rate',
                                  size=size, sizes=(20,2000), hue='country', alpha=0.5, legend=False)
   
    plt.axis((20, 100, 0, 10))

   
    plt.title('Life Expectancy vs Fertility' + str(year))
    plt.xlabel('Life Expectancy')
    plt.ylabel('fertility')
    plt.savefig('Life_Expectancy' + str(year) + '.png')

```
